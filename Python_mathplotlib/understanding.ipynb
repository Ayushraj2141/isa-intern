### Scatter Plot

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [4,6,8,11,4,2,9]

# colors =["r","y","g","b","r","g","r"] # for color -- konse colors available hai for changing ye bhi dekhnaa -- colors could also be changed as 
# a range of colors using cmap parameter -- colormap -- kaafi trh k hote hai with the flow chal jayegaa pataa
colors =[35,89,67,10,88,78,90]

# colorbar taaki pata chlee konsaa color use huaa haii

# want to change the sizes also ??
'''
Want to change to change the sizes too? -- do it using size parameter 
'''

# Want to change transperancy alsoo? isee alphaa parameter
# a = [0.1,0.2,0.40,0.50,0.20,1,0.70] -- Aise alag alag individually bhi change krr sktee haiii -- range from 0 to 1 only 
sizes = [1000,200,400,500,200,800,700]

'''
do not liked the plane circles? -- can change it using the marker syntax
aisaa nhi haii kuch bhi daal sktee hai ye dekhnaa pdegaa kya allowed hai and kya nhi 
jaise star allowed tha jab dekhaa but baaki kuch allowed nhi tha jaise @ , # ya a,b,c,d wgera wgeraa
edge color bhi change kr sktee haii using edgecolor ="color " -- thickness could also be change using linewidth = int
'''
plt.scatter(day,no, c=colors, s = sizes , alpha = 1 , marker = "*",edgecolors="r",linewidths=0.5,cmap="BrBG")
# Agar zyadaaa scatter functions usee krengee toh ekk hii plot k under bht saare points plot ho jayengee -- kaisee??
# plt.scatter(day,no2, c=colors, s = sizes , alpha = 0.8 ) -- could see by uncommenting this part 


# syntax with use of color or only c could also be used
# size nhi s use krnaa hai for sizes -- Also check ki ye size ki units kya haii?
# want to give title ?

t = plt.colorbar() # achaaa isme direct plt.colorbar.set_label krnee se nhi huaa kyunki woh uske properties mai nhi thaa shydd isse checkk krnaa ek baari
t.set_label("Color-Bar")
plt.title("Scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number",fontsize = 15)

plt.show()
```

### Bar Graph Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["Python","C++","Java"]
y = [90,60,70] # Popularity of these languages
z = [10,20,30] 
t = np.arange(len(x)) # iska use check krnaa
width = 0.8
p1 = [j + width for j in t]
# c = ["r","g","b"] # yaha pr c nhi color hi likhnaa pdegaa 
# align = senter / edge -- isse kahaa pr align krnaa hai 
plt.xlabel("Languages",fontsize = 20)
plt.ylabel("Popularity",fontsize = 20)
plt.title("Bar_graph-Plot")

# plt.bar(x,y,label = "plot" )#iske liye legend chahiyee  -- issee overwrite hogyaa ye nhi krnaa
plt.barh(p1,y,width, color = "r" , align = "center"  , alpha = 0.8) # for horizontal bargraph easyy just use -barh that's it 
plt.bar(p1,z,width, color = "y" , align = "center"  , alpha = 0.8 ) # but this will just overlap not distinguish 
#p1 ka use krkee side mai aajana chahiye thaa woh nhi aaya isse bhi ek baar check krnaa hai 

# plt.legend() -- iska usee krtee hai label lganne k liyee likhtee hai na graph mai 1 unit in x is equal to aise aisee 

plt.xticks(t+width,x,rotation = 10) # iske baare mai ek baar documentation mai pdhnaa 
plt.show()
```
